```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('inv/TRIPf.csv')
del df['% var.']
df.columns = ['date','close','open','high','low','volume']
df['date'] = pd.to_datetime(df['date'], format='%d.%m.%Y')
df = df.sort_values(by=['date'], ignore_index=True)
df.set_index(df['date'], inplace=True)
del df['date']
df
```

```python
for col in df.columns.tolist() :
    for i in range(len(df)) :
        value = df[col].values[i]
        value = value.replace(',','.')
        if col != 'volume' :
            value = float(value)
        else :
            last = value[-1]
            value = float(value[:-1])
            if last == 'K' :
                value *= 1000
            elif last == 'M' :
                value *= 1000000
        df[col].values[i] = value
    
df
```

```python
df.to_csv('TRIP.csv', sep=';')
```
