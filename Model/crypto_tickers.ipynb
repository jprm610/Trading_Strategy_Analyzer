# INVESTING

```python
driver = webdriver.Chrome('chromedriver.exe')

driver.get('https://www.investing.com/crypto/currencies')
```

```python
table = driver.find_element_by_tag_name("table")
rows = table.find_elements_by_tag_name("tr")
tickers = []
i = 0
for row in rows :
    i += 1
    cells = row.find_elements_by_tag_name("td")

    try :
        tickers.append(cells[3].text)
        print(i)
    except :
        continue
```

```python
tickers_df = pd.DataFrame({
    'ticker': np.array(tickers)
})
tickers_df.to_csv('crypto_tickers1.csv')
```

# YAHOO

```python
driver = webdriver.Chrome('chromedriver.exe')
```

```python
tickers = []
for i in range(0,400,100) :
    print(i)
    driver.get(f'https://finance.yahoo.com/cryptocurrencies/?count=100&offset={i}')
    time.sleep(10)
    table = driver.find_element_by_tag_name("table")
    rows = table.find_elements_by_tag_name("tr")
    j = 0
    for row in rows :
        j += 1
        cells = row.find_elements_by_tag_name("td")

        try :
            tickers.append(cells[0].text)
            print(j)
        except :
            continue
```

```python
tickers_df = pd.DataFrame({
    'tickers': np.array(tickers)
})
tickers_df.to_csv('cryptos_tickers.csv')
```
